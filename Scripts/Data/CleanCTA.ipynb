```python
CTA = pd.read_csv('CTA_-_System_Information_-_List_of__L__Stops.csv', encoding = 'gbk', header = 0)
```

```python
CTA.head()
```

```python
CTA.columns = CTA.columns.str.upper()
CTA['LATITUDE'] = 0
CTA['LONGITUDE'] = 0
CTA.head()
```

```python
for index,row in enumerate(CTA['LOCATION']):
    CTA['LATITUDE'][index] = row.replace('(','').replace(')','').split(',')[0]
    CTA['LONGITUDE'][index] = row.replace('(','').replace(')','').split(',')[1]

CTA.drop(['ADA', 'RED', 'BLUE', 'G', 'BRN', 'P', 'PEXP', 'Y', 'PNK', 'O'], axis = 1, inplace = True)
CTA.head()
```

```python
CTA.to_csv('CTA_cleaned.csv', encoding='utf-8', index=False)
```
